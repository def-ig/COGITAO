```python
transform_history = {}
for i in range(len(to_study)):
    transform = str(to_study[i]['transformation_suite'])
    if transform not in transform_history.keys():
        transform_history[transform] = 1
    else:
        transform_history[transform] += 1

for key in transform_history.keys():
    print(key, transform_history[key])
```
