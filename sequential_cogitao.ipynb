```python
# At each run, the generator will create a new task with a different combination of transformations, as described in the config file.
config = {
    "min_n_shapes_per_grid": 4,
    "max_n_shapes_per_grid": 4,
    "n_examples": 1,
    "min_grid_size": 20,
    "max_grid_size": 20,
    "allowed_combinations": [["mirror_vertical", "extend_contours_same_color", "empty_inside_pixels", 
                              "translate_up", "crop_top_side"]],
    "allowed_transformations": None,
    "min_transformation_depth": None,
    "max_transformation_depth": None,
    "shape_compulsory_conditionals": ["is_shape_less_than_10_rows", 
                                      "is_shape_less_than_10_cols"]
                                    
}

gen = generator.generator(config, debug_mode=True) # Initialize the generator with the config
task = gen.generate_single_task() # Generate a single task
```

```python
def recolor_temporarily_for_plotting(grid):
    new_grid = grid.copy()
    new_grid[grid == 7] = 3
    new_grid[grid == 5] = 2
    new_grid[grid == 8] = 1
    return new_grid

for i, grid in enumerate(task['full_grid_sequence']):
    # plot_grid(recolor_temporarily_for_plotting(grid), size = (10,10), save_path=f"paper_plots/grid_{i}.png")
    plot_grid(recolor_temporarily_for_plotting(grid), size = (1,1))
```
